```python
import pandas as pd
books=pd.read_csv('Books.csv',encoding="latin-1",delimiter=",")
books.head()
```

```python
ratings = pd.read_csv('Ratings.csv',encoding="latin-1",delimiter=",")
ratings.head()
```

```python
ratings_books=pd.merge(ratings,books,on='ISBN')
ratings_books.head()
```

```python
ratings_books.shape
```

```python
ratings_books.iloc[10000]
```

```python
ratings_books_sample=ratings_books.sample(frac=0.01,random_state=1)
ratings_books_sample.shape
ratings_books_sample.head(10)
```

```python
ratings_books_pivot=ratings_books_sample.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating').fillna(0)
ratings_books_pivot
```

```python
ratings_books_sample['Book-Title']
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(metric='cosine',algorithm='brute',n_neighbors=7,n_jobs=-1)
model.fit(ratings_books_pivot)
```

```python
indices=model.kneighbors(ratings_books_pivot.loc[["A Prayer for America"]],10,return_distance=False)[0]
```

```python
for index,values in enumerate(ratings_books_pivot.index[indices]):
  print((index+1),".",values)
```
